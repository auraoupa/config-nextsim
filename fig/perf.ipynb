```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib 

matplotlib.rc('xtick', labelsize=20) 
matplotlib.rc('ytick', labelsize=20) 
matplotlib.rcParams.update({'font.size': 20})

%matplotlib inline
```

```python
cores_1n=[4,8,16,32,64]
walltime_1n=[37,64,12,9,24]
cores_2n=[32]
walltime_2n=[30]
```

```python
marker_style = dict(linestyle=':', color='cornflowerblue', markersize=10)

fig, ax = plt.subplots()
fig.figsize=(20, 8)
plt.plot(cores_1n,walltime_1n,'.',marker='X',label='1 node',color='cornflowerblue')
plt.plot(cores_2n,walltime_2n,'.',marker='X',label='2 nodes',color='seagreen')

plt.xlabel('Number of cores')
plt.ylabel('Walltime in minutes')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Performance tests of neXtSIM on dahu - 10 days of simulation')
plt.savefig('perf-dahu-10d.png', bbox_inches='tight')
```
